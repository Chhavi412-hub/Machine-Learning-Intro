**Question 1: Explain the differences between AI, ML, Deep Learning (DL), and Data**
**Science (DS).**

**Here's a more detailed breakdown:**

This is the overarching concept of creating machines that can perform tasks that typically require human intelligence. AI can be rule-based or data-driven, and can encompass a wide range of tasks, from simple to complex.

**Machine Learning (ML):**
This is a subset of AI that focuses on enabling machines to learn from data without explicit programming. ML algorithms analyze data to identify patterns and make predictions or decisions.

**Deep Learning (DL):**
This is a specialized form of ML that utilizes artificial neural networks with multiple layers (deep neural networks) to analyze complex data. DL excels in tasks like image and speech recognition, natural language processing, and other areas where complex patterns are present.

**Data Science (DS):**
This is an interdisciplinary field that involves extracting knowledge and insights from data. Data scientists use various tools and techniques, including statistical analysis, data visualization, machine learning, and deep learning, to analyze data, identify trends, and solve problems. Data science emphasizes the entire data lifecycle, from data collection and cleaning to analysis and interpretation.

In essence, AI is the big picture, ML is a key technique for achieving AI, DL is a powerful method within ML, and DS is the broader field that utilizes these tools and techniques to gain insights from data

**Question 2: What are the types of machine learning? Describe each with one**
**real-world example.**

Ans Machine learning, a subset of AI, broadly categorizes into three main types: supervised, unsupervised, and reinforcement learning. Each type utilizes different approaches for learning from data to make predictions or decisions.

---



**1. Supervised Learning:**

**Description:**
This type involves training a model on a labeled dataset, where each data point has a corresponding correct output or label. The model learns to map inputs to outputs and can then predict outputs for new, unseen data.

**Example:**
Spam detection in emails. The model is trained on a dataset of emails labeled as either "spam" or "not spam". It learns patterns in the email content (words, sender, etc.) to classify new emails as spam or not spam.


---


**2. Unsupervised Learning:**

**Description:**
In unsupervised learning, the model learns from unlabeled data, finding hidden patterns, structures, or relationships within the data without explicit guidance.

**Example:**
Customer segmentation in marketing. An unsupervised learning algorithm can analyze customer data (purchase history, demographics, browsing behavior) to group customers into different segments with similar characteristics.

---



**3. Reinforcement Learning:**

**Description:**
This type involves training an agent to make decisions within an environment to maximize a cumulative reward. The agent learns through trial and error, receiving feedback in the form of rewards or penalties for its actions.

**Example:**
Training a self-driving car. The car (agent) learns to navigate a virtual environment (city streets) by receiving rewards for reaching its destination and penalties for collisions. Through this process, the car learns optimal driving strategies.

**Question 3: Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.**

Ans **Overfitting** occurs when a machine learning model learns the training data too well, capturing not only the underlying patterns but also noise and random fluctuations specific to that dataset. This leads to excellent performance on the training data but poor generalization and performance on new, unseen data. Overfit models typically exhibit low bias and high variance.

---



**Underfitting** occurs when a machine learning model is too simple to capture the underlying patterns in the data. It fails to learn the relationships between features and the target variable, resulting in poor performance on both the training data and unseen test data. Underfit models typically exhibit high bias and low variance.

---


**The Bias-Variance Tradeoff** is a fundamental concept in machine learning that describes the relationship between a model's complexity, its ability to capture true patterns (low bias), and its sensitivity to fluctuations in the training data (low variance).

---



**Bias**
refers to the error introduced by approximating a real-world problem, which may be complex, with a simplified model. High bias indicates that the model is too simple and makes strong assumptions about the data, leading to underfitting.

---



**Variance**
refers to the amount that the model's predictions vary when trained on different subsets of the training data. High variance indicates that the model is too sensitive to the specific training data, leading to overfitting.

**Question 4: What are outliers in a dataset, and list three common techniques for handling them.**

Ans Outliers in a dataset are data points that significantly deviate from the general pattern or distribution of the other data points. They are extreme values that are either much larger or much smaller than the majority of observations. Outliers can arise due to various reasons, including measurement errors, data entry mistakes, or genuine but rare occurrences within the data.

---



Three common techniques for handling outliers are:

**Removal/Deletion:**

This involves identifying and directly removing the outlier data points from the dataset. This method is straightforward but should be used cautiously, especially if the sample size is small or if the outliers represent valuable information. Common identification methods include using statistical thresholds like Z-scores (for normally distributed data) or the Interquartile Range (IQR) method (for skewed data).


---


**Transformation/Capping (Winsorization):**

This technique involves modifying the values of the outliers rather than removing them entirely. Winsorization, for instance, replaces outlier values with the nearest non-outlier value within a defined range (e.g., replacing values above the 99th percentile with the 99th percentile value). Other transformations like logarithmic or square root transformations can also be applied to reduce the influence of extreme values and make the data distribution more symmetrical.

---



**Robust Statistical Methods:**

Instead of modifying the data or removing outliers, this approach involves using statistical methods and models that are inherently less sensitive to the presence of outliers. Examples include using robust regression techniques (e.g., Huber regression, M-estimators) that down-weight the influence of outliers in parameter estimation, or utilizing non-parametric statistical tests which do not assume a specific data distribution and are thus more resilient to extreme values.


**Question 5: Explain the process of handling missing values and mention one**
**imputation technique for numerical and one for categorical data.**

Ans
**1. Deletion:**

**Rows:**
If a row has missing values, it can be removed entirely, especially if the missing data is significant or if the number of rows is large.

**Columns:**
If a column has a high proportion of missing values (e.g., more than half), it might be beneficial to drop the entire column to avoid skewing the analysis.

---



**2. Imputation:**

**Numerical Data:**

**Mean/Median Imputation:**
 Replace missing values with the mean or median of the respective column. This is a straightforward approach, but it can introduce bias if the data is not normally distributed or if there are outliers.

**K-Nearest Neighbors (KNN) Imputation:**
 KNN imputation estimates missing values based on the values of the K nearest neighbors in the dataset. This method is more sophisticated than mean/median imputation and can preserve data relationships.

**Categorical Data:**

**Most Frequent Category:** Replace missing values with the category that appears most frequently in the column. This is a simple and commonly used approach.

**Dedicated String:** Use a specific string like "Missing" or "Unknown" to represent missing values in the categorical column.

Question 8: Write a Python script to:

● Generate 1000 samples from a normal distribution.

● Introduce 50 random missing values.

● Fill missing values with the column mean.

● Plot a histogram before and after imputation.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate 1000 samples from a normal distribution
np.random.seed(42) # for reproducibility
data = np.random.normal(loc=0, scale=1, size=1000)
data_series = pd.Series(data)

# Introduce 50 random missing values
missing_indices = np.random.choice(len(data_series), size=50, replace=False)
data_with_missing = data_series.copy()
data_with_missing.iloc[missing_indices] = np.nan

# Plot histogram before imputation
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(data_with_missing.dropna(), bins=30, edgecolor='black')
plt.title('Histogram Before Imputation (Missing Values Removed)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Fill missing values with the column mean
mean_value = data_with_missing.mean()
data_imputed = data_with_missing.fillna(mean_value)

# Plot histogram after imputation
plt.subplot(1, 2, 2)
plt.hist(data_imputed, bins=30, edgecolor='black')
plt.title('Histogram After Mean Imputation')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Question 9: Implement Min-Max scaling on the following list of numbers [2, 5, 10, 15,

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

numbers = [2, 5, 10, 15, 20]

# Convert the list to a NumPy array and reshape for the scaler
numbers_array = np.array(numbers).reshape(-1, 1)

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Fit the scaler to the data and transform the data
scaled_numbers = scaler.fit_transform(numbers_array)

# Print the scaled array
print(scaled_numbers)
```
